```python
import numpy as np
import matplotlib.pyplot as plt

t = [0.0]
Logistic = [1.0]
Exponential = [1.0]
V_0 = 1
K = 1
a = 1
for i in range(0,5):
  Logistic.append(Logistic[i] + ((V_0*K)/(V_0+(K-V_0)*np.exp(-(a*i)))))
  Exponential.append(Exponential[i] + (V_0*np.exp(a*i)))
  t.append(i+1)

print("Exponential = ", Exponential, "\n")
print("Logistic = ", Logistic, "\n")
print("time = ",t,"\n")

#make figure
#set size of figure
plt.figure(figsize = (12, 8))
#plot time vs. volume
#'ko--' means color of line is black, markers are a circle, and want a dashed line
plt.plot(t,Exponential, 'ko--', label='Exponential')
plt.plot(t,Logistic, 'bo--', label='Logistic')

#make plot title
plt.title('Tumor Data from Table 1')
#label x-axis
plt.xlabel('Time (Days)')
#label y-axis
plt.ylabel('Volume ($mm^3$)')
#put grid lines on plot
plt.grid()
#show a legend
plt.legend(loc='lower right')
#display the plot
plt.show()
```
